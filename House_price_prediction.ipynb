```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
 #Create the dataset
data = {
    "Area": [100, 150, 200, 250, 300, 120, 180, 220, 270, 350],
    "Price": [30000, 40000, 50000, 60000, 70000, 32000, 42000, 52000, 62000, 75000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df=pd.read_csv('/content/houseprice.csv')
```

```python
df
```

```python
# Features (X) and target (y)
X = df[["Area"]]
y = df["Price"]

# Train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Display the coefficients and intercept
print("Coefficient (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)

```

```python
# Predict a value based on user input
area = float(input("Enter the area of the house (in sq ft): "))
predicted_price = model.predict([[area]])
print(f"Predicted Price for {area} sq ft is: Rs.{predicted_price[0]:.2f}")
```

```python
# Plot the data and the regression line
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, model.predict(X), color="red", label="Regression Line")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (Rs.)")
plt.title("House Price Prediction")
plt.legend()
plt.grid()
plt.show()
```
